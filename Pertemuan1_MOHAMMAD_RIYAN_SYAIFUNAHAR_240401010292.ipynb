## SEL 2 — Variabel dan Tipe Data Dasar

```python
nama     = "Mahasiswa Informatika"  # String  — teks
angkatan = 2024                     # Integer — bilangan bulat
ipk      = 3.75                     # Float   — bilangan desimal
aktif    = True                     # Boolean — True / False

print(f"Nama    : {nama}")
print(f"Angkatan: {angkatan}")
print(f"IPK     : {ipk}")
print(f"Aktif   : {aktif}")

print("\n--- Verifikasi Tipe Data ---")
print(f"type(nama)     = {type(nama)}")
print(f"type(angkatan) = {type(angkatan)}")
print(f"type(ipk)      = {type(ipk)}")
print(f"type(aktif)    = {type(aktif)}")
```

**Temuan SEL 2:**  
- Python memiliki 4 tipe data primitif utama: `str`, `int`, `float`, dan `bool`  
- Python bersifat **dynamically typed** — tipe data ditentukan otomatis saat variabel diberi nilai, tanpa deklarasi eksplisit  
- F-string (`f"...{variabel}..."`) adalah cara modern dan ringkas untuk menyisipkan nilai variabel ke dalam teks (Python >= 3.6)  
- Fungsi `type()` digunakan untuk mengecek tipe data suatu variabel

## SEL 3 — List dan Perulangan (Loop)

```python
tools = ["Python", "Google Colab", "Jupyter Notebook", "GitHub"]

print("Tools Data Science yang akan kita pelajari:")
for i, tool in enumerate(tools, 1):
    print(f"  {i}. {tool}")

print(f"\nJumlah tools  : {len(tools)}")
print(f"Tools pertama : {tools[0]}")
print(f"Tools terakhir: {tools[-1]}")
print(f"Slice [1:3]   : {tools[1:3]}")
```

**Temuan SEL 3:**  
- `list` adalah struktur data yang menyimpan kumpulan item secara berurutan dalam satu variabel  
- Indexing di Python dimulai dari **0** — `tools[0]` mengambil elemen pertama, `tools[-1]` mengambil elemen terakhir  
- `for ... in ...` digunakan untuk iterasi pada setiap elemen list satu per satu  
- `enumerate(list, start)` menghasilkan pasangan `(nomor_urut, nilai)` — praktis untuk tampilan bernomor  
- `len()` mengembalikan jumlah elemen dalam list

## SEL 4 — Fungsi (Function)

```python
def perkenalan(nama, prodi, angkatan):
    """Fungsi untuk menampilkan perkenalan mahasiswa."""
    return f"Hai! Saya {nama} dari {prodi}, angkatan {angkatan}."


hasil = perkenalan("Budi Santoso", "Informatika", 2024)
print(hasil)

hasil_saya = perkenalan("Mohammad Riyan Syaifunahar", "Informatika", 2024)
print(hasil_saya)
```

**Temuan SEL 4:**  
- Fungsi didefinisikan dengan kata kunci `def`, diikuti nama fungsi dan parameter dalam tanda kurung  
- **Docstring** (`"""..."""`) di baris pertama fungsi berfungsi sebagai dokumentasi — menjelaskan apa yang dilakukan fungsi  
- `return` mengembalikan nilai hasil fungsi ke pemanggil — tanpa `return`, fungsi mengembalikan `None`  
- Fungsi membuat kode lebih **modular dan reusable** — cukup tulis sekali, bisa dipanggil berkali-kali dengan input berbeda

## Eksplorasi — Menggabungkan Semua Konsep

```python
def tampilkan_profil(nama, nim, prodi, angkatan, tools_list):
    """
    Menampilkan profil lengkap mahasiswa beserta
    daftar tools Data Science yang akan dipelajari.
    """
    print("=" * 45)
    print("       PROFIL MAHASISWA DATA SCIENCE")
    print("=" * 45)
    print(f"  Nama     : {nama}")
    print(f"  NIM      : {nim}")
    print(f"  Prodi    : {prodi}")
    print(f"  Angkatan : {angkatan}")
    print("-" * 45)
    print("  Tools yang akan dipelajari:")
    for i, t in enumerate(tools_list, 1):
        print(f"    {i}. {t}")
    print("=" * 45)


tools_ds = ["Python", "Google Colab", "Jupyter Notebook", "GitHub"]

tampilkan_profil(
    nama       = "Mohammad Riyan Syaifunahar",
    nim        = "240401010292",
    prodi      = "Informatika",
    angkatan   = 2024,
    tools_list = tools_ds
)
```

**Temuan Eksplorasi:**  
- Variabel (`str`, `int`), list, perulangan `for` + `enumerate`, dan fungsi `def` dapat dikombinasikan untuk membuat program yang lebih bermakna  
- **Keyword argument** (`nama=..., nim=...`) membuat pemanggilan fungsi lebih jelas dan mudah dibaca  
- Pola ini — fungsi yang menerima data dan mencetak output terformat — adalah dasar dari hampir semua skrip Data Science
