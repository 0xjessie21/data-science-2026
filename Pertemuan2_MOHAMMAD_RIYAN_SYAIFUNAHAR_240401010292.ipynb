## Persiapan — Import Library

```python
import pandas as pd
import numpy as np
```

## 4.3 Pandas DataFrame — Latihan Dasar

```python
# Membuat DataFrame dari dictionary
data = {
    'nama': ['Ali', 'Budi', 'Caca', 'Dina'],
    'usia': [22, 25, 21, 23],
    'kota': ['Jakarta', 'Bandung', 'Surabaya', 'Medan'],
    'ipk' : [3.8, 3.5, 3.9, 3.7]
}
df_latihan = pd.DataFrame(data)

# Cek informasi dasar DataFrame
print(df_latihan.shape)    # jumlah baris & kolom
print(df_latihan.dtypes)   # tipe data tiap kolom
print(df_latihan.columns)  # nama kolom
df_latihan
```

## 4.4 Memuat Dataset Titanic dari URL

```python
# Load dataset dari URL GitHub
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Eksplorasi standar setelah load
df.head()     # 5 baris pertama
```

```python
df.info()      # tipe data & jumlah non-null per kolom
```

```python
df.describe()  # statistik: mean, std, min, max, dll
```

## 4.5 Filtering dan Seleksi Data

```python
# Filter penumpang yang selamat
selamat = df[df['Survived'] == 1]
print("Jumlah selamat:", len(selamat))

# Filter penumpang wanita
wanita = df[df['Sex'] == 'female']
print("Jumlah wanita:", len(wanita))
```

```python
# Filter wanita DAN kelas 1 (& = AND)
wanita_k1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print("Jumlah wanita kelas 1:", len(wanita_k1))
print(f"Rata-rata usia: {wanita_k1['Age'].mean():.2f} tahun")

# Lihat kolom tertentu saja
wanita_k1[['Name', 'Age', 'Survived']].head(10)
```

```python
# .loc[baris, kolom] — berbasis label
df.loc[df['Pclass'] == 1, ['Name', 'Age', 'Sex', 'Survived']].head()
```

## 4.6 Groupby dan Agregasi

```python
# Rata-rata Survived per kelas (pola dasar groupby)
df.groupby('Pclass')['Survived'].mean().round(3)
```

```python
# Agregasi multiple: total, selamat, % selamat, rata-rata usia
hasil = df.groupby('Pclass').agg(
    total       = ('PassengerId', 'count'),
    selamat     = ('Survived',    'sum'),
    pct_selamat = ('Survived',    'mean'),
    rata_usia   = ('Age',         'mean')
)
hasil['pct_selamat'] = (hasil['pct_selamat'] * 100).round(2)
hasil['rata_usia']   = hasil['rata_usia'].round(1)

print(hasil)

# Kelas mana yang paling tinggi survival rate-nya?
terbaik = hasil['pct_selamat'].idxmax()
print(f"\nKelas dengan survival tertinggi: Kelas {terbaik}")
```

```python
# Survival rate per kelas DAN jenis kelamin
df.groupby(['Pclass', 'Sex'])['Survived']\
  .mean().unstack().round(3)
```

## Q2 — Missing Values per Kolom

```python
# Jumlah missing values per kolom
missing = df.isnull().sum()

# Persentase missing values
persen = (df.isnull().sum() / len(df) * 100).round(2)

# Gabung menjadi tabel
missing_df = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persen
})

# Tampilkan hanya kolom yang punya missing values
print("Kolom dengan missing values:")
missing_df[missing_df['Jumlah Missing'] > 0]
```

**Temuan Q2:**  
Terdapat **3 kolom** dengan missing values:  
- `Age`: 177 data hilang (19.87%) — cukup signifikan, perlu strategi imputasi  
- `Cabin`: 687 data hilang (77.10%) — sangat banyak, sering diabaikan dalam analisis  
- `Embarked`: 2 data hilang (0.22%) — sangat sedikit, mudah diisi dengan modus

## Q3 — Persentase Penumpang yang Selamat

```python
# Menggunakan NumPy
survived_pct = np.mean(df['Survived']) * 100
print(f"Persentase selamat      : {survived_pct:.2f}%")
print(f"Persentase tidak selamat: {100 - survived_pct:.2f}%")
print(f"Jumlah selamat          : {df['Survived'].sum()} orang")
print(f"Jumlah tidak selamat    : {(df['Survived'] == 0).sum()} orang")

# Menggunakan Pandas (alternatif)
print("\nDistribusi Survived (Pandas):")
print((df['Survived'].value_counts(normalize=True) * 100).round(2))
```

**Temuan Q3:**  
Hanya **38.38%** (342 dari 891 penumpang) yang berhasil selamat.  
Lebih dari separuh penumpang yaitu **61.62%** (549 orang) tidak selamat dari tragedi Titanic.

## Q4 — Filter Penumpang Wanita Kelas 1

```python
# Filter boolean indexing: Sex == female DAN Pclass == 1
wanita_k1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

print(f"Jumlah penumpang wanita kelas 1 : {len(wanita_k1)} orang")
print(f"Rata-rata usia                  : {wanita_k1['Age'].mean():.2f} tahun")
print(f"Jumlah yang selamat             : {wanita_k1['Survived'].sum()} orang")
print(f"Survival rate                   : {wanita_k1['Survived'].mean()*100:.2f}%")

# Tampilkan 10 baris pertama
wanita_k1[['Name', 'Age', 'Survived']].head(10)
```

**Temuan Q4:**  
Terdapat **94 penumpang wanita di kelas 1** dengan rata-rata usia **34.61 tahun**.  
Kelompok ini memiliki survival rate sangat tinggi karena mendapat prioritas evakuasi sesuai kebijakan *"women and children first"*.

## Q5 — Tingkat Keselamatan per Kelas Penumpang

```python
# Groupby Pclass dengan agregasi lengkap
hasil = df.groupby('Pclass').agg(
    total       = ('PassengerId', 'count'),
    selamat     = ('Survived',    'sum'),
    pct_selamat = ('Survived',    'mean'),
    rata_usia   = ('Age',         'mean')
)
hasil['pct_selamat'] = (hasil['pct_selamat'] * 100).round(2)
hasil['rata_usia']   = hasil['rata_usia'].round(1)

print("Tingkat Keselamatan per Kelas Penumpang:")
print(hasil)

# Kelas dengan survival rate tertinggi
terbaik = hasil['pct_selamat'].idxmax()
print(f"\nKelas dengan survival tertinggi: Kelas {terbaik} ({hasil.loc[terbaik, 'pct_selamat']}%)")
```

**Temuan Q5:**  
**Kelas 1 (Pertama)** memiliki tingkat keselamatan tertinggi yaitu **62.96%**.  
Semakin rendah kelas penumpang, semakin kecil peluang selamat:  
- Kelas 1: 62.96% selamat  
- Kelas 2: 47.28% selamat  
- Kelas 3: 24.24% selamat  

Hal ini mencerminkan ketidaksetaraan akses ke sekoci penyelamat berdasarkan kelas sosial ekonomi.

## Q6 — Analisis Tambahan: Survival Rate Berdasarkan Jenis Kelamin dan Kelas

```python
# Survival rate per jenis kelamin dan kelas (groupby multi-kolom)
print("=== Survival Rate per Kelas & Jenis Kelamin (%) ===")
tabel = df.groupby(['Pclass', 'Sex'])['Survived']\
          .mean().unstack().round(3) * 100
print(tabel)

print("\n=== Survival Rate per Jenis Kelamin ===")
gender = df.groupby('Sex')['Survived'].agg(['sum', 'mean', 'count'])
gender.rename(columns={'sum': 'Selamat', 'mean': 'Rate', 'count': 'Total'}, inplace=True)
gender['Rate (%)'] = (gender['Rate'] * 100).round(2)
print(gender[['Total', 'Selamat', 'Rate (%)']])
```

**Temuan Q6:**  
**Jenis kelamin adalah faktor terkuat** yang mempengaruhi keselamatan penumpang Titanic:  
- Wanita kelas 1: **96.8%** selamat  
- Wanita kelas 2: **92.1%** selamat  
- Wanita kelas 3: **50.0%** selamat  
- Pria kelas 1: **36.9%** selamat  
- Pria kelas 2: **15.7%** selamat  
- Pria kelas 3: **13.5%** selamat  

Kombinasi **gender + kelas sosial** sangat menentukan peluang hidup.  
Ini mencerminkan dua faktor utama: (1) prioritas evakuasi *"women and children first"*, dan (2) ketimpangan akses ke dek sekoci berdasarkan kelas kabin.
